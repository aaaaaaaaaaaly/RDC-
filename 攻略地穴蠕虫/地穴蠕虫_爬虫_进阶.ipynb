```python
import re
import requests
from bs4 import BeautifulSoup
import urllib
import xlwt
```

## 获取数据

```python
def askurl(url,head):
    response=requests.get(url,headers=head)
    #救大命，编码乱了，查了csdn要这么写才行
    html=response.text.encode('iso-8859-1').decode('gbk')
    return html
```

## 对链接观察

```python
baseurl="https://www.zuowen8.com/a/"
#因为需要爬取10篇作文，随机试了一下，第二位1-10可以出现不同的作文网页，不过有一页不行，所以设置函数，跳过这一篇
for i in range(10):        
    i+=21
    url=baseurl+str(i)+"6265.html"
    #爬取
    print(url)

# 倒数第二个不行，就把那个链接的数字跳过，循环次数+1
```

## 定义函数

```python
def getdata (baseurl,header,data):
    for i in range(11):#因为要跳过一个链接，所以循环+1次
        i+=21
        if(i==29):#跳过没有作文的那一篇
            continue
        else:
            url=baseurl+str(i)+"6265.html"
            #爬取
            html=askurl(url,header)
            #解析
            soup=BeautifulSoup(html,"html.parser")
        
            #解析作文题目
            # item_1=soup.find_all('div',class_="artical")
            # item_1=str(item_1)
            findLink_1=re.compile(r'<h1>(.*?)</h1>')
            elements_1=re.findall(findLink_1,html)
    
            data.append(elements_1)
    
            #解析作文内容
            item_2=soup.find_all('div',class_="content")
            item_2=str(item_2)
            findLink_2=re.compile(r'<p>(.*?)</p>')
            elements_2=re.findall(findLink_2,item_2)
            #注意把elements转化成str(),不然会报错
            elements_2=re.sub(r'<u>.*?</u>.','',str(elements_2))
    
            data.append(elements_2)
```
